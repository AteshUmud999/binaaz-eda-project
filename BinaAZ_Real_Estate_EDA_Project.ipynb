# **Data Cleaning**

## Removing Identifier Columns

The columns `id_x`, `id_y`, `estate_id`, `estate_details_id_x`, and `estate_details_id_y` are unique identifiers assigned to each property record.

These columns are useful for database management and tracking individual records, but they do not contain meaningful information about property characteristics such as price, location, size, or other real estate features.

Since identifier columns do not contribute to exploratory data analysis and may introduce unnecessary complexity, they are removed from the dataset.


```python
df.drop(columns=[
    "id_x",
    "id_y",
    "estate_id",
    "estate_details_id_x",
    "estate_details_id_y"
], inplace=True)
```

```python
df.head()
```

## Removing URL and Image Reference Columns

The columns `img_url`, `rel_url`, `estate_rel_url`, `estate_rel_url_x`, and `estate_rel_url_y` contain website references and image links collected during the web scraping process.

These columns only provide access information to the original online listings and do not describe any property characteristics such as price, location, size, or other relevant attributes.

Since these features do not provide analytical value for exploratory data analysis, they are removed to reduce unnecessary complexity and keep the dataset focused on meaningful variables.


```python
df.drop(columns=[
    "img_url",
    "rel_url",
    "estate_rel_url",
    "estate_rel_url_x",
    "estate_rel_url_y"
], inplace=True)
```

```python
df.head()
```

## Removing Scraping Timestamp Columns

The columns `datetime_scrape_x` and `datetime_scrape_y` represent the timestamps when the data was collected during the web scraping process.

These columns describe the data extraction process rather than the actual characteristics of the properties. Since scraping timestamps do not provide meaningful information about property prices, locations, or features, they are removed from the dataset to focus on relevant analytical variables.


```python
df.drop(columns=[
    "datetime_scrape_x",
    "datetime_scrape_y"
], inplace=True)
```

```python
df.head()
```

The columns `currency_x` and `currency_y` contain the currency information for property prices.

After checking the unique values, it was observed that all records use the same currency (`AZN`). Since these columns contain no variation and do not provide additional information for exploratory data analysis, they are removed from the dataset.

```python
df["currency_x"].value_counts(dropna=False)
```

```python
df.drop(columns=[
    "currency_x",
    "currency_y"
], inplace=True)
```

## Removing Non-Analytical Columns

### Removing `description`

The `description` column contains free-text information written by property owners or agents. Although this data could be useful for Natural Language Processing (NLP) tasks, it is not suitable for this exploratory data analysis because it consists of unstructured text with a high number of unique values.

Since the analysis focuses on structured property features such as price, location, size, and property characteristics, the `description` column is removed to reduce complexity and improve dataset readability.

---

### Removing `owner_name`

The `owner_name` column contains personal names of property owners or listing providers. This information does not provide meaningful insights into property characteristics, pricing patterns, or market behavior.

Since it represents personal identification information rather than an analytical feature, the `owner_name` column is removed from the dataset.

---

### Removing `updated`

The `updated` column contains information about when a property listing was last updated on the website.

This information represents website activity and data collection details rather than property-related characteristics. Since it does not contribute to understanding property prices or features, the `updated` column is removed.


```python
df.drop(columns=[
    "description",
    "owner_name",
    "updated"
], inplace=True)
```

```python
df.head()
```

## Removing Secondary Date and Time Columns

The columns `day_y` and `hour_y` represent additional date and time information collected from another source during the scraping process.

After comparing them with `day_x` and `hour_x`, it was observed that these columns contain many missing values and do not provide additional consistent information. The `Binanın növü` column contains an extremely high percentage of missing values, making it unreliable for meaningful analysis.
