### Converting Land Area from Sot to Square Meters

The `Sahə` column contains area values in different units:
- Square meters (`m²`)
- Sot (`sot`)

For consistency in analysis, all area values are converted into square meters.
Since 1 sot equals 100 square meters, sot values are multiplied by 100.

```python
import numpy as np

def convert_area(value):
    if pd.isna(value):
        return np.nan

    value = str(value).lower().strip()

    if "sot" in value:
        number = float(value.replace("sot", "").strip())
        return number * 100

    elif "m²" in value:
        number = float(value.replace("m²", "").strip())
        return number

    return np.nan


df["Sahə"] = df["Sahə"].apply(convert_area)
```

```python
df.head()
```

### Cleaning Unit Price Feature

The `unit_price` column contains price per square meter values stored as text with currency symbols.

The currency symbol and extra characters are removed, and the column is converted into a numerical format for further analysis.

```python
df["unit_price"] = (
    df["unit_price"]
    .str.replace(" AZN/m²", "", regex=False)
    .str.replace(" ", "", regex=False)
)

df["unit_price"] = pd.to_numeric(
    df["unit_price"],
    errors="coerce"
)
```

### Filling Missing Unit Price Values

Missing unit prices are calculated using the relationship between total property price and area.

Formula:

Unit Price = Total Price / Area

This method preserves the actual property characteristics instead of replacing values with statistical estimates.

```python
df["unit_price"] = df["unit_price"].fillna(
    df["price"] / df["Sahə"]
)
```

# **Outlier Detection and Handling**

Outliers were identified using the Interquartile Range (IQR) method.

The IQR method was selected because property prices and areas are typically not normally distributed and may contain extreme values. It is more robust to skewed distributions than the z-score method. **kalın metin**

```python
df.head()
```

```python
numerical_cols = [
    "price",
    "unit_price",
    "Sahə",
    "views"
]
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    percentage = outliers / len(df) * 100

    print(f"{col}: {outliers} ({percentage:.2f}%)")
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print("\n", col)
    print("-------------------")
    print(f"Lower bound: {lower_bound:.2f}")
    print(f"Upper bound: {upper_bound:.2f}")
    print(f"Minimum value: {df[col].min()}")
    print(f"Maximum value: {df[col].max()}")
```

# Feature Engineering

In this step, new meaningful features were created from existing columns to improve the quality of analysis and extract additional insights from the real estate dataset.

The newly created features are based on domain knowledge and existing property characteristics.

### 1. Price Category

A new feature called `price_category` was created based on the property price.

The purpose of this feature is to group properties into different price segments, making it easier to analyze the distribution of affordable, medium, expensive, and luxury properties.

Price categories were defined based on domain knowledge:
- Low: Properties with lower prices
- Medium: Mid-range properties
- High: Expensive properties
- Luxury: Very high-value properties

```python
def price_category(price):
    if price < 100000:
        return "Low"
    elif price < 300000:
        return "Medium"
    elif price < 1000000:
        return "High"
    else:
        return "Luxury"


df["price_category"] = df["price"].apply(price_category)
```

```python
df.head()
```

### 2. Area Per Room

A new feature called `area_per_room` was created by dividing the total property area by the number of rooms.

This feature represents the average living space per room and helps to better understand property layouts and compare different types of apartments.

```python
df["area_per_room"] = np.where(
    (df["Otaq sayı"] > 0) & (df["Sahə"] <= 1000),
    df["Sahə"] / df["Otaq sayı"],
    np.nan
)
```

### 📊 Price Distribution
- **Log Scale Applied:** Transformed to handle high right-skewness caused by high-end properties.
- **Key Insight:** Most property prices cluster around the median value, while a small percentage of luxury listings stretch the upper tail.

```python
import matplotlib.ticker as ticker


sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))

ax = sns.histplot(
    data=df,
    x="price",
    bins=50,
    kde=True,
    log_scale=True,
    color="#1f77b4",
    edgecolor="white",
    alpha=0.7,
)

median_val = df["price"].median()
mean_val = df["price"].mean()

plt.axvline(
    median_val,
    color="#d62728",
    linestyle="--",
    linewidth=2,
    label=f"Median: {median_val:,.0f} AZN",
)
plt.axvline(
    mean_val,
    color="#2ca02c",
    linestyle="-.",
    linewidth=1.5,
    label=f"Mean: {mean_val:,.0f} AZN",
)


def price_formatter(x, pos):
    if x >= 1e6:
        return f"{x*1e-6:g}M"
    elif x >= 1e3:
        return f"{x*1e-3:g}K"
    else:
        return f"{x:g}"


ax.xaxis.set_major_formatter(ticker.FuncFormatter(price_formatter))

plt.title("Overall Price Distribution", fontsize=13, fontweight="bold")
plt.xlabel("Price (AZN)", fontsize=11)
plt.ylabel("Number of Listings", fontsize=11)
plt.legend(frameon=True, facecolor="white")
plt.grid(True, alpha=0.3)

plt.show()
```
