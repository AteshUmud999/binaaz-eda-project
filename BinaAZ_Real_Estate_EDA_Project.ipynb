### Handling Missing Values in VIP Feature

The `vip` column contains information about premium advertisements.
Missing values mean that the property does not have a VIP promotion.

Instead of removing these rows, missing values are replaced with `not_vip` because the absence of a value has a meaningful interpretation.

```python
df["vip"] = df["vip"].fillna("not_vip")
```

### Handling Missing Values in Featured Feature

The `featured` column indicates whether a listing is highlighted as a featured advertisement.

Null values represent properties without featured status, so they are replaced with `not_featured`.

```python
df["featured"] = df["featured"].fillna("not_featured")
```

### Handling Missing Values in Products Label

The `products_label` column represents additional advertising packages.

Missing values indicate that the listing does not use any additional product or promotion.
Therefore, null values are replaced with `no_label`.

```python
df["products_label"] = df["products_label"].fillna("no_label")
```

### Converting Land Area from Sot to Square Meters

The `Sahə` column contains area values in different units:
- Square meters (`m²`)
- Sot (`sot`)

For consistency in analysis, all area values are converted into square meters.
Since 1 sot equals 100 square meters, sot values are multiplied by 100.

```python
import numpy as np

def convert_area(value):
    if pd.isna(value):
        return np.nan

    value = str(value).lower().strip()

    if "sot" in value:
        number = float(value.replace("sot", "").strip())
        return number * 100

    elif "m²" in value:
        number = float(value.replace("m²", "").strip())
        return number

    return np.nan


df["Sahə"] = df["Sahə"].apply(convert_area)
```

```python
df.head()
```

### Cleaning Unit Price Feature

The `unit_price` column contains price per square meter values stored as text with currency symbols.

The currency symbol and extra characters are removed, and the column is converted into a numerical format for further analysis.

```python
df["unit_price"] = (
    df["unit_price"]
    .str.replace(" AZN/m²", "", regex=False)
    .str.replace(" ", "", regex=False)
)

df["unit_price"] = pd.to_numeric(
    df["unit_price"],
    errors="coerce"
)
```

### Filling Missing Unit Price Values

Missing unit prices are calculated using the relationship between total property price and area.

Formula:

Unit Price = Total Price / Area

This method preserves the actual property characteristics instead of replacing values with statistical estimates.

```python
df["unit_price"] = df["unit_price"].fillna(
    df["price"] / df["Sahə"]
)
```

```python
df.head()
```

# **Outlier Detection and Handling**

Outliers were identified using the Interquartile Range (IQR) method.

The IQR method was selected because property prices and areas are typically not normally distributed and may contain extreme values. It is more robust to skewed distributions than the z-score method. **kalın metin**

```python
df.head()
```

```python
numerical_cols = [
    "price",
    "unit_price",
    "Sahə",
    "views"
]
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    percentage = outliers / len(df) * 100

    print(f"{col}: {outliers} ({percentage:.2f}%)")
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print("\n", col)
    print("-------------------")
    print(f"Lower bound: {lower_bound:.2f}")
    print(f"Upper bound: {upper_bound:.2f}")
    print(f"Minimum value: {df[col].min()}")
    print(f"Maximum value: {df[col].max()}")
```

### Outlier Threshold Analysis

Using the IQR method, upper boundaries were calculated for numerical features.

- Properties with prices above 627,500 AZN were considered potential price outliers.
- Unit prices above 4,309.76 AZN/m² were identified as potential outliers.
- Property areas above 295 m² were considered extreme values.
- View counts above 1,551 were detected as potential outliers.

Some extreme values, especially in price and views, may represent real high-value properties or popular listings. Therefore, outliers were not removed automatically and were kept for further analysis.

```python
for col in ["price", "unit_price", "Sahə", "views"]:
    print("\n", col)
    display(df.nlargest(5, col)[[col, "location", "Kateqoriya", "Otaq sayı", "address"]])
```

```python
for col in ["price", "unit_price", "Sahə"]:
    print("\n", col)
    display(df.nsmallest(5, col)[[col, "location", "Kateqoriya", "address"]])
```

```python
# Price outliers
price_outliers = df[
    (df["price"] < 1000) |
    (df["price"] > 10000000)
]

print("Price outliers:", len(price_outliers))


# Unit price outliers
unit_price_outliers = df[
    (df["unit_price"] < 100) |
    (df["unit_price"] > 10000)
]

print("Unit price outliers:", len(unit_price_outliers))


# Area outliers
area_outliers = df[
    (df["Sahə"] < 5) |
    (df["Sahə"] > 5000)
]

print("Sahə outliers:", len(area_outliers))
```
